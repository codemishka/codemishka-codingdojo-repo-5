```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
# Assuming your data file is named 'medical_data.csv'
df = pd.read_csv('medical_data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Plotting histogram with KDE for specified features
plt.figure(figsize=(12, 8))

# Features to analyze
features = ['VitD_levels', 'Doc_visits', 'TotalCharge']

for feature in features:
    plt.subplot(2, 2, features.index(feature) + 1)
    sns.histplot(df[feature], kde=True, color='skyblue', bins=30)

    # Add lines for mean, median, and +/- 1 std
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    std_val = df[feature].std()

    plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(mean_val + std_val, color='black', linestyle='dashed', linewidth=2, label='+1 Std')
    plt.axvline(mean_val - std_val, color='black', linestyle='dashed', linewidth=2, label='-1 Std')

    # Highlight the range between +1 and -1 std
    plt.axvspan(mean_val - std_val, mean_val + std_val, color='yellow', alpha=0.3, label='+/- 1 Std Range')

    # Set plot labels and title
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Histogram with KDE for {feature}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Features to analyze
features = ['VitD_levels', 'Doc_visits', 'TotalCharge']

# Set the style for the plots
sns.set(style="whitegrid")

# Plot histograms with kde and add lines for mean, median, +1 std, -1 std
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[feature], kde=True, color='blue', bins=30)

    # Add lines for mean, median, +1 std, -1 std
    plt.axvline(df[feature].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(df[feature].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(df[feature].mean() + df[feature].std(), color='black', linestyle='dashed', linewidth=2, label='+1 Std')
    plt.axvline(df[feature].mean() - df[feature].std(), color='black', linestyle='dashed', linewidth=2, label='-1 Std')

    # Highlight the range between +1 and -1 std
    plt.axvspan(df[feature].mean() - df[feature].std(), df[feature].mean() + df[feature].std(), color='yellow', alpha=0.3, label='Â±1 Std Range')

    # Set plot labels and title
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature}')
    
    # Show legend
    plt.legend()

    # Show the plot
    plt.show()
    
# Answer the questions
print(f"\nAnalysis for {feature}:")
print(f"Is it Discrete or Continuous? {['Continuous', 'Discrete'][df[feature].nunique() <= 10]}")

# Check skewness
skewness = df[feature].skew()
skewness_direction = 'No skew' if skewness == 0 else ('Positive skew' if skewness > 0 else 'Negative skew')
print(f"Does it have a skew? If so, which direction (+/-): {skewness_direction}")

# Check kurtosis
kurtosis = df[feature].kurtosis()
kurtosis_type = 'Mesokurtic' if kurtosis == 0 else ('Leptokurtic' if kurtosis > 0 else 'Platykurtic')
print(f"What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic): {kurtosis_type}")

print("\n" + "=" * 50 + "\n")
```
