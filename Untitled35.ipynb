```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import missingno as msno
import seaborn as sns
import numpy as np
```

```python
fname='/content/drive/MyDrive/CodingDojo/04-DataEnrichment/Week15/Data/data (4) (1).csv'
df = pd.read_csv(fname)
df.info()
df.head()
```

```python


# Plotting histogram with KDE for specified features
plt.figure(figsize=(12, 8))

# Features to analyze
features = ['VitD_levels', 'Doc_visits', 'TotalCharge']

for feature in features:
    plt.subplot(2, 2, features.index(feature) + 1)
    sns.histplot(df[feature], kde=True, color='skyblue', bins=30)

    # Add lines for mean, median, and +/- 1 std
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    std_val = df[feature].std()

    plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(mean_val + std_val, color='black', linestyle='dashed', linewidth=2, label='+1 Std')
    plt.axvline(mean_val - std_val, color='black', linestyle='dashed', linewidth=2, label='-1 Std')

    # Highlight the range between +1 and -1 std
    plt.axvspan(mean_val - std_val, mean_val + std_val, color='yellow', alpha=0.3, label='+/- 1 Std Range')

    # Set plot labels and title
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Histogram with KDE for {feature}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Descriptive statistics for each feature
for feature in features:
    print(f"\nDescriptive Statistics for {feature}:")
    print(df[feature].describe())

    # Check skewness
    skewness = df[feature].skew()
    print(f"Skewness: {skewness}")

    # Check kurtosis
    kurtosis = df[feature].kurtosis()
    print(f"Kurtosis: {kurtosis}")

    print("\n" + "-" * 40)

```

Mean: Represents the average value of the feature.
Skewness: Measures the asymmetry of the distribution. Negative skewness indicates a longer tail to the left, while positive skewness indicates a longer tail to the right. A value close to zero suggests symmetry.
Kurtosis: Describes the tails of the distribution. Positive kurtosis (leptokurtic) indicates heavier tails, while negative kurtosis (platykurtic) indicates lighter tails. A value close to zero suggests a mesokurtic distribution, which is approximately normal.
