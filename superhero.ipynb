```python
df_info[['Hero','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
## drop the original column
df_info = df_info.drop(columns=['Hero|Publisher'])
df_info.head(2)
```

```python
## use .str.replace to replace all single quotes
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
## examining a single value from the coordinates col
measurements = df_info.loc[0,"Measurements"]
print(type(measurements))
measurements
```

```python
height_weight = df_info['Measurements'].apply(pd.Series)
height_weight
```

```python
# concat measurements with original dataframe
df_info = pd.concat((df_info, height_weight), axis = 1)
df_info.head(2)
```

```python
df_info = df_info.drop(columns=['Measurements'])
df_info.head(2)
```

```python
#Remove 'cm' from height column
df_info[['Height','Units']] = df_info['Height'].str.split(' ',expand=True)
df_info.head(2)
```

```python
#Remove 'kg' from weight column
df_info[['Weight','Units2']] = df_info['Weight'].str.split(' ',expand=True)
df_info.head(2)
```

```python
# drop the Unit columns
df_info = df_info.drop(columns=['Units', 'Units2'])
df_info.head(2)
```

```python
df_info['Height'] = pd.to_numeric(df_info['Height'])
df_info['Weight'] = pd.to_numeric(df_info['Weight'])
df_info.info()
```

```python
#One-Hot encode the comma separated values
df_power = pd.concat([df_power.drop('Powers', 1), df_power['Powers'].str.get_dummies(sep=",")], 1)
# check results
df_power.head()
```

```python
#merge the 2 dataframes into one
df_super = pd.merge(df_info, df_power, left_on='Hero', right_on='hero_names')
df_super.head()
```

```python
#Compare the average weight of super powers who have Super Speed to those who do not.
avg_with_superspeed = df_super[df_super['Super Speed'] == 1]['Weight'].mean()
avg_without_superspeed = df_super[df_super['Super Speed'] == 0]['Weight'].mean()

print ('Average weight of superpowers with Super Speed: ',  avg_with_superspeed)
print ('Average weight of superpowers without Super Speed: ',  avg_without_superspeed)
```

```python

#What is the average height of heroes for each publisher?.
avg_height_per_publisher = df_super.groupby('Publisher')['Height'].mean()

print('Average height of heroes per publisher: ', avg_height_per_publisher)
```
